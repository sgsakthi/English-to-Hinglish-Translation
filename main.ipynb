```python
# Read the content of the text file
with open("english.txt", "r") as file:
    file_contents = file.readlines()

# Find the position of the "task" subheading
task_position = None
for index, line in enumerate(file_contents):
    if "task" in line.lower():
        task_position = index
        break

# Extract points based on specific format and handle multi-line points
points = []
point_content = ""

for line in file_contents[task_position+1:]:
    # Checking if line starts with a point indicator (like "1. ", "2. ", "3. ")
    if any(line.startswith(f"{i}. ") for i in range(1, 4)):
        # If we already have content in point_content (from previous points), we add it to points list
        if point_content:
            points.append(point_content)
            point_content = ""
        point_content += line.strip()
    # If line doesn't start with a point indicator but has content, we add it to point_content
    elif line.strip() and not any(line.startswith(f"{i}. ") for i in range(1, 4)):
        point_content += " " + line.strip()
    # If we find an empty line, we add current point_content to points and reset point_content
    elif not line.strip() and point_content:
        points.append(point_content)
        point_content = ""
    
    # Breaking the loop if we've collected all 3 points
    if len(points) == 3:
        break

# If there's still content in point_content after the loop, we add it to points
if point_content:
    points.append(point_content)

points_to_convert = points[1:4]

# Print the extracted points
for point in points_to_convert:
    print(point)
```

```python
# Convert the list to a single string with each item on a new line
content = '\n'.join(points_to_convert)

# Write the content to a file name 'Extracted English output.txt'
with open('Extracted English output.txt', 'w') as file:
    file.write(content)
```

```python
import openai
import os
from dotenv import load_dotenv, find_dotenv

# Load environment variables from .env file
load_dotenv(find_dotenv())

# Set OpenAI API key from environment variable
openai.api_key = os.environ.get("OPENAI_API_KEY")


def translate_to_hinglish(english_text):
    """
    Translate English text to Hinglish using OpenAI API.
    """
    response = openai.Completion.create(
        engine="davinci",
        prompt=f"Translating English text into casual Hinglish while retaining certain English terms for clarity: '{english_text}'",
        max_tokens=500  # Set a limit to avoid overly long responses
    )
    
    return response.choices[0].text.strip()

# Read the content of the uploaded Extracted English output.txt file
with open("Extracted English output.txt", "r") as file:  
    english_content = file.read()

# Translate to Hinglish
hinglish_translation = translate_to_hinglish(english_content)

# Save to a new file
with open("hinglish_output1.txt", "w") as output_file:
    output_file.write(hinglish_translation)
```

```python
## Back to PDF (Optional)
from fpdf import FPDF

def txt_to_pdf_updated(txt_file_path, pdf_file_path):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=10)
    
    with open(txt_file_path, 'r', encoding='utf-8') as file:
        for line in file:
            # Encode line into Latin-1, replace problematic characters with a space instead of '?'
            pdf.cell(150, 7, txt=line.encode('latin-1', 'replace').decode('latin-1').replace('?', ' '), ln=True)
    
    pdf.output(pdf_file_path)


txt_to_pdf_updated("hinglish_output1.txt", "output_in_Hinglish1.pdf")
```
